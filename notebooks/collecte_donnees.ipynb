## Collecte de données
Le but du projet étant de mettre en place un modèle de prédictions des résultats de match de foot, nous allons dans un premier temps collecter les résultats des matchs des 6 dernières saisons de Premier League (Championnat Anglais)

```python
import pandas as pd

seasons = {
    "2018-2019": "1819",
    "2019-2020": "1920",
    "2020-2021": "2021",
    "2021-2022": "2122",
    "2022-2023": "2223",
    "2023-2024": "2324"
}

all_matches = []

for season, code in seasons.items():
    url = f"https://www.football-data.co.uk/mmz4281/{code}/E0.csv"
    print(f"Downloading {season}")

    df = pd.read_csv(url)
    df["season"] = season
    all_matches.append(df)

matches_df = pd.concat(all_matches, ignore_index=True)

matches_df.to_csv(
    "../data/raw/premier_league_matches_raw.csv",
    index=False
)

matches_df.head()
```

## Nettoyage de données :
Dans ce notebook, le but sera de nettoyer les données pour une meilleure lisibilité ainsi que pour leur utilisation futur par nos modèles de prédiction.

```python
import pandas as pd
import os
```

```python
df = pd.read_csv("/content/premier_league_matches_raw.csv")
df.head()
```

```python
df.columns
```

Nous allons renommer les colonnes pour plus de lisibilité et garder uniquement celles qui nous intéressent.

```python
dic = {
    "Date": "date_match",
    "HomeTeam": "equipe_dom",
    "AwayTeam": "equipe_ext",

    "FTHG": "buts_dom",
    "FTAG": "buts_ext",
    "FTR": "resultat",

    "HS": "tirs_dom",
    "AS": "tirs_ext",
    "HST": "tirs_cadres_dom",
    "AST": "tirs_cadres_ext",

    "HY": "jaunes_dom",
    "AY": "jaunes_ext",
    "HR": "rouges_dom",
    "AR": "rouges_ext"
}

df = df.rename(columns=dic)

```

```python
colonnes_a_garder = [
    "date_match",
    "equipe_dom",
    "equipe_ext",

    "buts_dom",
    "buts_ext",
    "resultat",

    "tirs_dom",
    "tirs_ext",
    "tirs_cadres_dom",
    "tirs_cadres_ext",

    "jaunes_dom",
    "jaunes_ext",
    "rouges_dom",
    "rouges_ext",

    "cible_resultat",
    "season"
]
df = df[colonnes_a_garder]
```

```python
df["cible_resultat"] = df["resultat"].map({
    "H": 2,
    "D": 1,
    "A": 0
})
```

```python
df[["resultat", "cible_resultat"]].value_counts()
```

```python
df.to_csv(
    "/content/premier_league_nettoye.csv",
    index=False
)
```

Bien évidemment ceci n'est qu'une approche préliminaire, le but final serait de prédire le résultat d'un match en fonction de plus de paramètres. Mais il faudrait commencer par une analyse simpliste.
